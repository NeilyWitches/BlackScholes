```python
import csv
import locale
import math
import pandas as pd
import numpy as np
```

```python
def BlackScholes(file_path, K, r, T):
    results = []
    locale.setlocale(locale.LC_ALL, 'en_US.UTF8')
    
    with open(file_path) as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            results.append(row)
            
    S = [None]*(len(results) - 1)
    date = [None]*(len(results) - 1)
    for i in range(1,len(results)):
        S[i-1] = locale.atof(results[i][1].strip(" $"))
        date[i-1] = results[i][0]
        
        
    S.reverse()
    date.reverse()
        
    mu = sum(S)/len(S)
    i = 0
    list = [None]*len(S)
    t = [None]*len(S)
    
    for i in range(len(S)):
        list[i] = (S[i] - mu)**2
        t[i] = i
        
    var = sum(list)/len(S)
    std = math.sqrt(var)
    #std = 0.8
    
    d1 = [None]*len(S)
    d2 = [None]*len(S)
    a = 0
    dx = 0.1
    xrange1 = [None]*len(S)
    xrange2 = [None]*len(S)
    n1 = [None]*len(S)
    n2 = [None]*len(S)
    x1 = [None]*len(S)
    x2 = [None]*len(S)
    summ1 = [None]*len(S)
    summ2 = [None]*len(S)
    y1 = [None]*len(S)
    y2 = [None]*len(S)
    N1 = [None]*len(S)
    N2 = [None]*len(S)
    F = [None]*len(S)
    i = 0
    
    for i in range(len(S)):
        d1[i] = (math.log(S[i]/K) + (r + 1/2*var)*(T - t[i]))/(std*math.sqrt(T - t[i]))
        d2[i] = d1[i] - std*math.sqrt(T - t[i])
    
        xrange1[i] = d1[i] - a
        xrange2[i] = d2[i] - a
        n1[i] = int(xrange1[i]/dx)
        n2[i] = int(xrange2[i]/dx)
        x1[i] = [None]*n1[i]
        x2[i] = [None]*n2[i]
        summ1[i] = 0
        summ2[i] = 0
        for j in range(n1[i]):
            x1[i][j] = a + j*dx
            y1[i] = 1/math.sqrt(2*math.pi)*math.exp(-1/2*x1[i][j]**2)
            summ1[i] = summ1[i] + y1[i]*dx
        for k in range(n2[i]):
            x2[i][k] = a + j*dx
            y2[i] = 1/math.sqrt(2*math.pi)*math.exp(-1/2*x2[i][k]**2)
            summ2[i] = summ2[i] + y2[i]*dx
            
        N1[i] = 0.5 + summ1[i]
        N2[i] = 0.5 + summ2[i]
        
        F[i] = round(S[i]*N1[i] - K*math.exp(-r*(T - t[i]))*N2[i],2)
        
    np_array = np.transpose([date, S, F])
    new_file = file_path.replace('.csv', 'withcalloptionprices2.csv')
    
    return pd.DataFrame(np_array).to_csv(new_file, header = ["Time", "Stock Price", "Call Option Price"], index = None)
        
    
 
    
    
```
